# Logistic Regression - Tunning Hyperparameters

```python
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV, train_test_split
import pandas as pd
import numpy as np
```

```python
ds = datasets.load_breast_cancer()
print(ds.keys())
filename = ds.filename
cols = ds.feature_names
cols = [i.replace(' ','_') for i in cols]
cols.append('breast_cancer')
print(cols)
```

```python
breast = pd.read_csv(filename, header=None, names=cols)
breast.head()
```

```python
breast.info()
```

```python
breast.dropna(subset=['breast_cancer'], inplace=True)
breast['mean_perimeter'] = breast['mean_perimeter'].astype('float')
breast['mean_area'] = breast['mean_area'].astype('float')
breast.info()
```

```python
X = breast.drop('breast_cancer', axis=1)
y = breast['breast_cancer']
c_space = np.logspace(-5, 8, 15)
param_grid = {'C': c_space, 'penalty':['l1','l2']}
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)
```

```python
logreg = LogisticRegression(solver='liblinear')
logreg_cv = GridSearchCV(logreg, param_grid, cv=5)
logreg_cv.fit(X_train, y_train)
```

```python
print("Tuned Logistic Regression Parameters: {}".format(logreg_cv.best_params_)) 
print("Best score is {}".format(logreg_cv.best_score_))
```
